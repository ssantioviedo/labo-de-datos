Esta forma de ir modificando un dataset no es muy práctica.

Por ejemplo si ejecutamos varias veces el mismo comando, obtendremos cada vez algo distinto.

Una solución es encadenar métodos para hacer el código más legible y fácil de reusar. Y más eficiente también.

```python
penguinsClean = penguins.dropna().reset_index()
penguinsClean.head()
```

```python
# Si queremos eliminar la columna de índices viejos
penguinsClean = penguins.dropna().reset_index(drop = True)
penguinsClean.head()
```

## 3. Funciones lambda
Resulta engorroso definir una función para usarla solamente dentro del rename.

Las funciones lambda nos permiten definir funciones anónimas "al vuelo".

La sintaxis es

`lambda arguments : expression`

```python
# Ejemplo
x = lambda a : a + 10
print(x(5))
```

```python
# Ejemplo
x = lambda a, b : a * b
print(x(5, 6))
```

```python
# Ejemplo
x = lambda a : a + 1
print(x(5))
```

```python
# Ahora sí!
penguinsClean.rename(index = lambda a : a + 1)
```

Esta operación la podemos encadenar con las anteriores:

```python
penguinsClean = penguins.dropna().reset_index(drop = True).rename(index = lambda a : a + 1)
penguinsClean.head()
```

```python
# O si queremos cada metodo en una linea nueva
penguinsClean = (
    penguins.dropna()
    .reset_index(drop = True)
    .rename(index = lambda a : a + 1)
)
penguinsClean.head()
```

**Ejercicio:** Encadanar un método más para agregar "_orig" a los nombres de todas las columnas.

```python
# O si queremos cada metodo en una linea nueva
penguinsClean = (
    penguins.dropna()
    .reset_index(drop = True)
    .rename(index = lambda a : a + 1)
    .rename(columns = lambda a : a + "_orig")
)
penguinsClean.head()
```

## Extra: pipes
Una cadena de operaciones encadenadas suele llamarse un "pipe" (tubería).

Por ejemplo, en R podemos hacer pipes en un DataFrame con %>%:

`first(x) %>% second(x) %>% third(x)`

En Python podemos encadenar métodos directamente poniendo uno a continuación del otro, por lo tanto en general no necesitamos un comando especial para pipes.

Igualmente, los DataFrames de Pandas también tienen un método `pipe()` para encadenar funciones aplicadas al DataFrame que puede resultar útil.

```python
# Ejemplo
def sumaUnoIndex(x):  # Esta función recibi un DataFrame
    x.index += 1 # Le sumamos uno al indice
    return(x)
```

```python
penguinsClean = (
    penguins.dropna()
    .reset_index(drop = True)
    .rename(columns = lambda a : a + "_orig")
    .pipe(sumaUnoIndex)  # Reemplamos el rename() por este pipe
)
penguinsClean.head()
```

### Aplicar funciones distintas según alguna condición.

La función que usamos para listas por comprensión es una función lambda `(lambda x : f(x))`, pero solo ponemos f(x).

La funciones lambda permite usar condiciones if de la siguiente forma para definir funciones partidas

`lambda x : f(x) if CONDICION else g(x)`

C. Para listas por comprensión con entonces usamos la sintaxis

`[f(x) if CONDICION else g(x) for x in LISTA if CONDICION]`


```python
# Ejemplo. Qué hace el siguiente comando?
[x / 2 if (x % 2 == 0) else x for x in np.arange(1,10)]   # np.arange nos da todos los enteros desde 1 a 10 (exclusive)
```
